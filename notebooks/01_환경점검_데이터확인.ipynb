````python
from pathlib import Path
import sys
import platform
import importlib.util

# ── 루트 경로 설정 ──────────────────────────────────────────
ROOT = Path.cwd()
# notebooks/ 폴더에서 실행된 경우 한 단계 올라감
if not (ROOT / 'scripts').exists() and (ROOT.parent / 'scripts').exists():
    ROOT = ROOT.parent

DATA_DIR   = ROOT / 'data'
IMAGE_DIR  = DATA_DIR / 'images'
VIDEO_DIR  = DATA_DIR / 'videos'
OUTPUT_DIR = ROOT / 'outputs'
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print('Python:', sys.version.replace('\n', ' '))
print('Platform:', platform.platform())
print('워크스페이스 루트:', ROOT)
print('이미지 폴더:', IMAGE_DIR)
print('동영상 폴더:', VIDEO_DIR)
````

````python
# 패키지 점검
for name in ['ultralytics', 'cv2', 'matplotlib', 'pandas', 'PIL']:
    status = 'OK' if importlib.util.find_spec(name) else 'MISSING ← pip install 필요'
    print(f'{name}: {status}')
````

````python
# 이미지 / 동영상 파일 목록 확인
image_exts = {'.jpg', '.jpeg', '.png', '.bmp'}
video_exts = {'.mp4', '.avi', '.mov', '.wmv', '.mkv'}

images = sorted([p for p in IMAGE_DIR.iterdir() if p.is_file() and p.suffix.lower() in image_exts]) \
         if IMAGE_DIR.exists() else []
videos = sorted([p for p in VIDEO_DIR.iterdir() if p.is_file() and p.suffix.lower() in video_exts]) \
         if VIDEO_DIR.exists() else []

print('[이미지 샘플]')
if images:
    for p in images:
        print(' -', p.name)
else:
    print(' - 없음 → data/images/ 에 .jpg/.png 파일을 넣어주세요.')

print('\n[동영상 샘플]')
if videos:
    for p in videos:
        print(' -', p.name)
else:
    print(' - 없음 → data/videos/ 에 .mp4 파일을 넣어주세요.')
    print(' ※ Google Colab 환경에서는 .wmv 재생이 불가합니다. .mp4를 권장합니다.')
````
